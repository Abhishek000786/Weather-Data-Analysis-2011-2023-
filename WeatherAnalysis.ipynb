```python
import pandas as pd

# 1. Average Wind Speed Analysis
def average_wind_speed_analysis(df):
    wind_speed_columns = ['WindSpeed9am', 'WindSpeed3pm']
    avg_wind_speeds = df[wind_speed_columns].mean()
    
    print("Average Wind Speeds:")
    print(avg_wind_speeds,'\n')
    

# 2. Average Humidity Analysis
def average_humidity_analysis(df):
    humidity_columns = ['Humidity9am', 'Humidity3pm']
    avg_humidities = df[humidity_columns].mean()
    
    print("Average Humidities:")
    print(avg_humidities,'\n')

# 3. Average Pressure Analysis
def average_pressure_analysis(df):
    pressure_columns = ['Pressure9am', 'Pressure3pm']
    avg_pressures = df[pressure_columns].mean()
    
    print("Average Pressures:")
    print(avg_pressures,'\n')

# 4. Average Temperature Analysis
def average_temperature_analysis(df):
    temp_columns = ['Temp9am', 'Temp3pm', 'MinTemp', 'MaxTemp']
    avg_temperatures = df[temp_columns].mean()
    
    print("Average Temperatures:")
    print(avg_temperatures,'\n')

# 5. Rainfall Analysis
def rainfall_analysis(df):
    avg_rainfall = df['Rainfall'].mean()
    total_rainfall = df['Rainfall'].sum()
    max_rainfall = df['Rainfall'].max()
    min_rainfall = df['Rainfall'].min()
    
    print(f"Average Rainfall: {avg_rainfall:.2f} mm")
    print(f"Total Rainfall: {total_rainfall:.2f} mm")
    print(f"Max Rainfall: {max_rainfall:.2f} mm")
    print(f"Min Rainfall: {min_rainfall:.2f} mm")

# Perform EDA
average_wind_speed_analysis(df)
average_humidity_analysis(df)
average_pressure_analysis(df)
average_temperature_analysis(df)
rainfall_analysis(df)
```

```python
import pandas as pd

# Define the function to handle outliers
def handle_outliers(df, column):
    # Calculate Q1, Q3, IQR, and bounds
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identify outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    
    # Return bounds, outliers, and the number of outliers
    return lower_bound, upper_bound, outliers, len(outliers)

# Apply outlier handling to relevant columns
columns_to_check = ['Temp9am', 'Temp3pm', 'Rainfall', 'WindGustSpeed']
for column in columns_to_check:
    lower_bound, upper_bound, outliers, num_outliers = handle_outliers(df, column)
    
    print(f"Column: {column}")
    print("Number of Outliers:", num_outliers)
    print(f"Lower Bound: {lower_bound}")
    print(f"Upper Bound: {upper_bound}")
    print(f"Outliers in {column}:\n{outliers}\n")
```
